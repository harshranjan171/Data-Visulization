```python
from matplotlib import pyplot as plt
```

```python
x =[1, 2, 3]
y =[1, 4, 9]
plt.plot(x, y)
plt.show()
```

```python
x =[1, 2, 3]
y =[1, 4, 9]
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x =[1, 2, 3]
y =[1, 4, 9]
z=[10 , 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.show()
```

```python
x =[1, 2, 3]
y =[1, 4, 9]
z=[10 , 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["THIS IS y" , "THIS IS z"])
plt.show()
```
